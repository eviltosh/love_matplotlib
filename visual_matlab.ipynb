```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data=[[15,4],[70,32],[100,49]]
```

```python
df_family=pd.DataFrame(data,index=["babybear","mommabear","pappabear"],columns=["Weight in Kilo","Age"])
```

```python
df_family
```

```python
def graph_example():
    graph=input("What type of graph bar line or pie?")
    if graph=="pie":
        df_family['Weight in Kilo'].plot(kind="pie")
        plt.show()
    elif graph=="bar":
        df_family.plot(kind=graph)
        plt.show()
    elif graph=="line": 
        df_family.plot(kind=graph)
        plt.show()
    else:
        print("good_bye")
    
```

```python
while True:
    graph_example()
    response=input("Do you wish to see more? ENTER yes")
    if response != "yes":
        break
```
